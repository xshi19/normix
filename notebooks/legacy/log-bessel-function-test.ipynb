```python
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.optimize import minimize
from scipy.linalg import solve_triangular
from scipy.special import kv

sys.path.append('../')

from pygh.gig import GIG
from pygh.func import logkv, kvratio

%matplotlib inline
```

```python
plt.style.use('seaborn')
```

```python
for v, z_min, z_max in [(10, 1e-2, 1), (100, 1e-2, 1e3), (200, 1e-2, 1e3), (500, 1e-2, 1e3)]:
    z = np.linspace(1e-2, z_max, 10000)
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(z, logkv(v, z), 'b', label='log kv')
    ax.plot(z, np.log(kv(v, z)), 'r', label='scipy kv')
    ax.legend()
    ax.set_title('v='+str(v))
```
