```python
from scipy.special import gammaln
```

```python
def logkv(v, z):
    y = np.log(kv(v, z))
    
    if np.any(y==-np.inf):
        
    
    if np.any(y==np.inf):
        if np.abs(v) > 1e-10:
            y[y==np.inf] = gammaln(np.abs(v)) + np.abs(v) * (np.log(2.) - np.log(z[y==np.inf])) - np.log(2.)
        else:
            y[y==np.inf] = np.log(-np.log(z[y==np.inf]/2) - np.euler_gamma)
            
    return y
```
